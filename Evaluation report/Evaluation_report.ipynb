```python
# p-value Pearsons correlation
def pearsons_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 50)
    return pvalues
```

```python
# Pearsons correlation
def pearsons_correlation(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    corr_coeff = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            corr_coeff[r][c] = round(pearsonr(df[r], df[c])[0], 50)
    return corr_coeff
```

```python
# p-value Spearmanr correlation
def spearmanr_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(spearmanr(df[r], df[c])[1], 50)
    return pvalues
```

```python
# Spearmanr correlation
def spearmanr_correlation(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    corr_coeff = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            corr_coeff[r][c] = round(spearmanr(df[r], df[c])[0], 50)
    return corr_coeff
```

```python
# p-value Kendalltau correlation
def kendalltau_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(stats.kendalltau(df[r], df[c])[1], 50)
    return pvalues
```

```python
# Kendalltau correlation
def kendalltau_correlation(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    corr_coeff = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            corr_coeff[r][c] = round(stats.kendalltau(df[r], df[c])[0], 50)
    return corr_coeff
```

```python
def bonf_correction(df):
  p_adjusted = multipletests(df, method='bonferroni')
  return p_adjusted[1] #1 is for corrected p-values for bonferroni method
```
